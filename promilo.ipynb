```python
#Traffic_Aquisition

Traffic_Aqu = pd.read_excel('Data set for BA (2).xlsx', sheet_name = 2)
Traffic_Aqu
```

```python
Traffic_Aqu.info()
```

```python
Traffic_Aqu.describe()
```

```python
plt.figure(figsize=(7,5))
plt.bar(Traffic_Aqu['Session default channel group'], Traffic_Aqu['Users'], color='khaki')
plt.xlabel('Session default channel group')
plt.ylabel('Users')
plt.title('Users on Traffic Aquisition')
plt.xticks(rotation=45)
plt.show()
```

```python
cmp0=User_aqu.groupby('First user default channel group')['New users'].sum()
cmp1=Traffic_Aqu.groupby('Session default channel group')['Users'].sum()
fig, axes= plt.subplots(1,2,figsize=(9,5))
axes[0].bar(cmp0.index, cmp0.values, color='lightcoral')
axes[0].set_xlabel('Session default channel group')
axes[0].set_ylabel('Users')
axes[0].set_title('Users on User Acquisition')
axes[0].tick_params(axis='x', labelrotation=90)

axes[1].bar(cmp1.index, cmp1.values, color='peachpuff')
axes[1].set_xlabel('First user default channel group')
axes[1].set_ylabel('Users')
axes[1].set_title('Users on Traffic Aquisition')
axes[1].tick_params(axis='x', labelrotation=90)
plt.tight_layout()
plt.show()
```

```python
#Google Ads Report

g_ads = pd.read_excel('Data set for BA (2).xlsx', sheet_name=14)
g_ads
```

```python
g_ads.info()
```

```python
g_ads.describe()
```

```python
ads=g_ads.groupby('Session Google Ads campaign')['Google Ads clicks'].sum().sort_values(ascending=True)
ads1=g_ads.groupby('Session Google Ads campaign')['Google Ads cost per click'].sum().sort_values(ascending=True)
fig, axes = plt.subplots(1,2,figsize=(12,6))
axes[0].barh(ads.index, ads.values, color = 'gold')
axes[0].set_xlabel('Clicks')
axes[0].set_ylabel('Session Google Ads campaign')
axes[0].set_title('Ads clicks by Google Ads campaign')

axes[1].barh(ads1.index, ads1.values, color ='greenyellow')
axes[1].set_xlabel('Cost per click')
axes[1].set_ylabel('Session Google Ads campaign')
axes[1].set_title('Cost per click by Google Ads campaign')
plt.tight_layout()
plt.show()
```
